```python
#Applying statistical operations on the Car_dataset

import statistics
import pandas as pd 
# reading the data by read_csv command
car_dataset=pd.read_csv("C:\Desktop\AI&ML-LAB\car_price_dataset.csv")

#finding the mean of the price coloumn of car_dataset
mean_price= car_dataset['Price'].mean()
print("Median of Price:", mean_price)

#finding the median of the price coloumn of car_dataset
median_price= car_dataset['Price'].median()
print("Median of Price:", median_price)

#finding the mode of the price coloumn of car_dataset
mode_price= car_dataset['Price'].mode()[0] 
print("Mode of Price:", mode_price)

#finding the Standard deviation of the price coloumn of car_dataset
standard_deviation_price= car_dataset['Price'].std()
print("Standard Deviation of Price:", standard_deviation_price)
```

```python
# Checking the Distribution of 'Price' of Car_dataset

import matplotlib.pyplot as plt #visualization library
plt.hist(car_dataset['Price'], bins=20, color='navy', edgecolor='white')
plt.xlabel("Car Price")
plt.ylabel("Frequency")
plt.title("The Price distribution of Cars")
plt.show()
```

```python
#checking Skewness and kurtosis of 'Price' of Car_dataset

from scipy.stats import skew, kurtosis

#skewness batati ki data symmetrical hai ya nahi
#Agar Skewness > 0, tou prices right-skewed hain (zyada mehengi hain cars).
skewness= skew(car_dataset['Price'])
print(f"Skewness of Car Prices: {skewness}")

#kurtosis batati hai ki data me outliers ziyada hain ki nahi
#Kurtosis > 3 ho to outliers zyada hain, aur < 3 ho to normal hai.
kurt= kurtosis(car_dataset['Price'])
print(f"Kurtosis of Car Prices: {kurt}")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

# Reading dataset
car_dataset = pd.read_csv("C:/Desktop/AI&ML-LAB/car_price_dataset.csv")

# Extracting Price column
price_data = car_dataset['Price']

# Calculating Mean & Standard Deviation
mean_price = price_data.mean()
std_price = price_data.std()

# Plot Histogram
plt.figure(figsize=(8,5))
sns.histplot(price_data, bins=20, kde=False, color='navy', edgecolor='white', stat='density')

# Generate Normal Curve
x = np.linspace(price_data.min(), price_data.max(), 100)
y = norm.pdf(x, mean_price, std_price)

# Overlay Normal Curve
plt.plot(x, y, color='red', label="Normal Distribution Curve")

# Labels and Title
plt.xlabel("Car Price")
plt.ylabel("Density")
plt.title("Price Distribution of Cars with Normal Curve")
plt.legend()

# Show plot
plt.show()
```
